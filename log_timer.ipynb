### We create a function to return the execution time

```python
def tim_func(function,text):
	t1 = time()
	result = function(text)
	t2 = time()
	return (t1-t2)
```

```python
list_dict=[]
list_counter=[]
for i in range(10):
    list_dict.append(tim_func(count_word2,words))
    list_counter.append(tim_func(count_word_3,words))
```

```python
import matplotlib.pyplot as plt
fig=plt.figure()
plt.plot(range(len(list_counter)),list_counter,label="counter")
plt.plot(range(len(list_dict)),list_dict,label="dict")
plt.xlabel("iterations")
plt.ylabel("exe_time")
plt.legend()
plt.title("Comparison of execution time between two type of counting words")
plt.show()
```
